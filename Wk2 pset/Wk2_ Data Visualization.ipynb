```python
import numpy as np
from matplotlib import pyplot as plt
from scipy import misc as mc
import pandas as pd
import seaborn as sns
from matplotlib.colors import LogNorm
```

Question #1

```python
def prob1():
    data = np.load('anscombe.npy')
    Ix = list(data[:,0])
    Iy = list(data[:,1])
    IIx = list(data[:,2])
    IIy = list(data[:,3])
    IIIx = list(data[:,4])
    IIIy = list(data[:,5])
    IVx = list(data[:,6])
    IVy = list(data[:,7])
    x1 = np.linspace(0,20)

    plt.subplot(221)
    plt.plot(Ix,Iy,'o',color='k',markersize=3)
    plt.plot(x1,0.5*x1+3)
    plt.tight_layout()

    plt.subplot(222)
    plt.plot(IIx,IIy,'o',color='k',markersize=3)
    plt.plot(x1,0.5*x1+3)
    plt.tight_layout()

    plt.subplot(223)
    plt.plot(IIIx,IIIy,'o',color='k',markersize=3)
    plt.plot(x1,0.5*x1+3)
    plt.tight_layout()

    plt.subplot(224)
    plt.plot(IVx,IVy,'o',color='k',markersize=3)
    plt.plot(x1,0.5*x1+3)
    plt.tight_layout()

    return plt.show()

prob1()
print("It is interesting to note the differences between each graph. While the first appears to be a correlated scatter, the remaining graphs appear to be strongly correlated with a function (despite having near-identical core properties at first glance).")
```

Question #2

```python
def Bernstein(x,n,v):
    fx = mc.comb(n,v) * (x**v) * ((1.0 - x)**(n-v))
    return fx

def prob2(m):
    xaxis = list(np.linspace(0,1,100))
    for i in range(1,m+2):
        for j in range(1,i+1):
            yaxis = [Bernstein(x,i-1,j-1) for x in xaxis]
            plt.subplot(m+1,m+1,j + 4*(i-1))
            plt.plot(xaxis,yaxis)
            plt.title("n = "+str(i-1)+", v = "+str(j-1))
            plt.tick_params(which="major", right="off", left="off")
            plt.tight_layout()
            plt.axis([0,1,0,1])
    return plt.show()

prob2(3)
```

Question #3

```python
def prob3():
    data = np.load("MLB.npy")
    height = data[:,0]
    weight = data[:,1]
    age = data[:,2]

    m1,b1 = np.polyfit(height,weight,1)
    plt.subplot(221)
    plt.plot(height,m1*height + b1)
    plt.plot(height,weight,"o", markersize=2)
    plt.tight_layout
    plt.title("Height and Weight")

    m2,b2 = np.polyfit(height,age,1)
    plt.subplot(222)
    plt.plot(height,m2*height + b2)
    plt.plot(height,age,"o", markersize=2)
    plt.tight_layout
    plt.title("Height and Age")

    m3,b3 = np.polyfit(weight,age,1)
    plt.subplot(223)
    plt.plot(weight,m3*weight + b3)
    plt.plot(weight,age,"o", markersize=2)
    plt.tight_layout
    plt.title("Weight and Age")

    return plt.show()

prob3()
```

Question #4

```python
import numpy as np
import matplotlib.pyplot as plt
def prob4():
    data = np.load("earthquakes.npy")
    year = data[:,0]
    magnitude = data[:,1]
    longitude = data[:,2]
    latitude = data[:,3]

    plt.hist(year, bins=11, color="r")
    plt.tight_layout()
    plt.title("Number of Earthquakes by Year")
    plt.xlabel("Year")
    plt.ylabel("Frequency")
    plt.show()

    plt.hist(magnitude, bins=20, color="r")
    plt.tight_layout()
    plt.title("Frequency of Earthquakes by Magnitude")
    plt.xlabel("Magnitude")
    plt.ylabel("Frequency")
    plt.show()

    plt.scatter(longitude,latitude,c=magnitude,marker="o",cmap="inferno",edgecolors="face")
    plt.title("Location of Earthquakes by Magnitude")
    plt.xlabel("Longitude")
    plt.ylabel("Latitude")
    plt.show()

    return plt.show()

prob4()
```

Question #5

```python
def Rosenbrock(x,y):
    fxy = (1 - x)**2 + 100*(y - x**2)**2
    return fxy

def prob5():
    axis = np.linspace(-15,15)
    X,Y = np.meshgrid(axis,axis)
    plt.pcolormesh(X,Y,Rosenbrock(X,Y), cmap="inferno", norm=LogNorm())
    plt.scatter([1],[1], color="w")
    return plt.show()

prob5()
```

Problem #6

```python
country_names = ["Austria","Bolivia","Brazil","China","Finland","Germany","Hungary","India","Japan","North Korea","Montenegro","Norway","Peru","South Korea","Sri Lanka","Switzerland","Turkey","United Kingdom","United States","Vietnam"]

def prob6(nations):
    pos = range(len(nations))
    data = np.load("countries.npy")
    pop = data[:,0]
    gdp = data[:,1]
    h_m = data[:,2]
    h_f = data[:,3]

    plt.plot(pop,gdp,".")
    plt.tight_layout()
    plt.title("GDP and Population")
    plt.xlabel("Population, millions")
    plt.ylabel("2015 GDP, billions USD")
    plt.show()

    plt.barh(pos,gdp,align="center")
    plt.yticks(pos,nations)
    plt.tight_layout
    plt.title("GDP by Country")
    plt.xlabel("2015 GDP, billions USD")
    plt.show()

    plt.hist(h_m, bins=10)
    plt.tight_layout()
    plt.title("Male Heights, Frequency")
    plt.xlabel("Height Intervals, centimeters")
    plt.ylabel("Frequency")
    plt.show()

    plt.hist(h_f, bins=10)
    plt.tight_layout()
    plt.title("Female Heights, Frequency")
    plt.xlabel("Height Intervals, centimeters")
    plt.ylabel("Frequency")
    plt.show()

    return plt.show()

prob6(country_names)

print("A number of useful insights can be gleaned from these graphs. First, is clear to see that as GDP per capita is quite variable amongst the countries tested, we do not have a terribly strong correlation. In addition, female heights are not as well distributed as female heights amongst nations.")
```
