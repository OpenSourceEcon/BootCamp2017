```python
import numpy as np
from matplotlib import pyplot as plt
```

Question #1

```python
def ar_gen(mu,s,n):
    arr = []
    for i in range(n):
        arr.append(np.random.normal(mu,s,n))
    return np.array(arr)

K = ar_gen(0,1,1000)
var_K = np.var(K, axis=0)
plt.plot(K)
plt.show()
```

Question #2

```python
def prob2():
    xax = np.linspace(-2*np.pi, 2*np.pi, 50000)
    sin_x = [np.sin(x) for x in xax]
    cos_x = [np.cos(x) for x in xax]
    arctan_x = [np.arctan(x) for x in xax]
    plt.plot(xax, sin_x)
    plt.plot(xax, cos_x)
    plt.plot(xax, arctan_x)
    return plt.show()

prob2()
```

Problem #3

```python
def f(x):
    fx = 1 / (x - 1)
    return fx

def prob3():
    xax = np.linspace(-2,6,100)
    yax1 = [f(x) for x in xax if x < 1]
    yax2 = [f(x) for x in xax if x > 1]
    xax1 = [x for x in xax if f(x) < 0]
    xax2 = [x for x in xax if f(x) > 0]
    plt.plot(xax1,yax1,'m',linewidth=4)
    plt.plot(xax2,yax2,'m',linewidth=4)
    plt.ylim(-6,6)
    return plt.show()

prob3()
```

Question #4

```python
def prob3():
    xax = np.linspace(0, 2*np.pi,10000)

    plt.subplot(2,2,1)
    plt.plot(xax,np.sin(xax))
    plt.axis([0,2*np.pi,-2,2])
    plt.title("sinx")

    plt.subplot(2,2,2)
    plt.plot(xax,np.sin(2*xax))
    plt.axis([0,2*np.pi,-2,2])
    plt.title("sin2x")

    plt.subplot(2,2,3)
    plt.plot(xax,2*np.sin(xax))
    plt.axis([0,2*np.pi,-2,2])
    plt.title("2sinx")

    plt.subplot(2,2,4)
    plt.plot(xax,2*np.sin(2*xax))
    plt.axis([0,2*np.pi,-2,2])
    plt.title("2sin2x")

    plt.tight_layout()
    plt.suptitle("Signs")
    return plt.show

prob3()
```

```python
def prob5():
    data = np.load("FARS.npy")
    hour = list(data[:,0])
    longitude = list(data[:,1])
    latitude = list(data[:,2])

    plt.subplot(211)
    plt.plot(longitude, latitude, 'o', color='k', markersize=0.1)
    plt.axis("equal")
    plt.xlabel("Longitude")
    plt.ylabel("Latitude")
    plt.tight_layout()

    plt.subplot(212)
    plt.hist(hour, bins=np.arange(0,24), color='m')
    plt.xlabel("Hour")
    plt.tight_layout()

    return plt.show()

prob5()
```

Problem #6

```python
def f(x,y):
    fxy = (np.sin(x)*np.sin(y))/(x*y)
    return fxy

def prob6():
    xaxis = np.linspace(-2*np.pi,2*np.pi)
    X,Y = np.meshgrid(xaxis,xaxis)

    plt.subplot(121)
    plt.pcolormesh(X,Y,f(X,Y),cmap='prism')
    plt.colorbar()

    plt.subplot(122)
    plt.contour(X,Y,f(X,Y),20,cmap='prism')
    plt.colorbar()

    return plt.show()

prob6()
```
